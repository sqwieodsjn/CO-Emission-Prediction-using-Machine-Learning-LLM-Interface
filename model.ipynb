```python
import joblib
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder,RobustScaler
```

```python
df =pd.read_csv('first_project.csv')
```

```python
df = df.drop(columns=['Make','Model','Vehicle Class'])
```

```python
df['Transmission'].unique()
```

```python
x = df.drop(columns=['CO2 Emissions(g/km)'])
y = df['CO2 Emissions(g/km)']
```

```python
cat_cols = x.select_dtypes(include=['object']).columns
num_cols = x.select_dtypes(include=['int64','float64']).columns
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
ordinal = OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1)
x_train[cat_cols] = ordinal.fit_transform(x_train[cat_cols])
x_test[cat_cols] = ordinal.transform(x_test[cat_cols])
```

```python
scalar = RobustScaler()
x_train[num_cols] = scalar.fit_transform(x_train[num_cols])
x_test[num_cols] = scalar.transform(x_test[num_cols])
```

```python
x_train
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
joblib.dump(ordinal, 'encoder.pkl')
joblib.dump(scalar, 'scaler.pkl')
joblib.dump(model, 'model.pkl')
```
